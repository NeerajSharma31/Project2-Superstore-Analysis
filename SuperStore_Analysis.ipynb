```python
df_recency =df.groupby(by='Customer Name',
                      as_index=False)['Order Date'].max() # group of "customer name" column 
df_recency.columns = ['CustomerName','LastPurchaseDate']
recent_date = df_recency['LastPurchaseDate'].max()

df_recency['Recency']=df_recency['LastPurchaseDate'].apply(
lambda x: (recent_date - x).days)

df_recency.head()
```

```python
df_recency['LastPurchaseDate'].max() 
```

```python
frequency_df =df.drop_duplicates().groupby(
by=['Customer Name'],as_index= False)['Order Date'].count()  


frequency_df.columns = ['CustomerName', 'Frequency']  # changed name of column "order date" to "Frequency" and Customer Name to "CustomerName"
frequency_df.head()


# frequency tells about the how many time customers purchased
```

```python
df['Total']=df['Sales']*df['Quantity']
monetary_df = df.groupby ( by= 'Customer Name', as_index=False)['Total'].sum()

monetary_df.columns=['CustomerName','Monetary']

monetary_df.head()


#MONETARY CALCUALTE THE TOTAL AMOUNT PURCHACED BY CUSTOEMR 
```

```python
rf_df = df_recency.merge(frequency_df, on= 'CustomerName')
rfm_df = rf_df.merge(monetary_df, on='CustomerName').drop(
columns='LastPurchaseDate')
rfm_df.head()

# 
```

```python
rfm_df['R_rank']=rfm_df['Recency'].rank(ascending=False)  
rfm_df['F_rank']=rfm_df['Frequency'].rank(ascending=True) 
rfm_df['M_rank']=rfm_df['Monetary'].rank(ascending=True) 


# normalizing the rank of the customers
rfm_df['R_rank_norm'] = (rfm_df['R_rank']/rfm_df['R_rank'].max())*100
rfm_df['F_rank_norm'] = (rfm_df['F_rank']/rfm_df['F_rank'].max())*100
rfm_df['M_rank_norm'] = (rfm_df['M_rank']/rfm_df['M_rank'].max())*100

# we have normalized so it brings all data in one unit and so we  can delete it.

rfm_df.drop(columns=['R_rank','F_rank','M_rank'], inplace=True)

rfm_df.head()
```

```python
rfm_df['RFM_Score']=0.15*rfm_df['R_rank_norm']+0.28*\
rfm_df['F_rank_norm']+0.57*rfm_df['M_rank_norm']

rfm_df['RFM_Score']*=0.05  #by multiply 0.05 it will be shift decimal and range will reduce in numbers and easy to understand

rfm_df=rfm_df.round(2)
rfm_df[['CustomerName', 'RFM_Score']].head(20)

#rfm_df['RFM_Score'].min() 
```

```python
rfm_df['RFM_Score'].describe()
```



#now here is whos score is above 4 so they all are priority customers and very imp for company
# RATINGS CUSTOMER BASED UPON THE RFM SCORE:
* RFM SCORE>4.5: TOP CUSTOEMRS
* 4.5 > RFM SCORE >4 : HIGH VALUE CUSTOMERS
* 4> RFM SCORE >3: MEDIUM VALUE CUSTOMER
* 3> RFM SCORE> 1.5 : LOW VALUE CUSTOEMR
* RFM SCORE <1.5 : LOST CUSTOMER

```python
rfm_df['Customer_segment']= np.where(rfm_df['RFM_Score']> 4.5,
                                    "Top Customers",
                                    (np.where(
                                    rfm_df['RFM_Score']>4,
                                    "High Value Customer",
                                    (np.where(
                                    rfm_df['RFM_Score']>3,
                                    "Medium Value Customer",
                                     np.where(rfm_df['RFM_Score']>1.5,
                                             "Low Value Customers", 'Lost Customers'))))))

rfm_df[['CustomerName','RFM_Score','Customer_segment']].head(20)
```
